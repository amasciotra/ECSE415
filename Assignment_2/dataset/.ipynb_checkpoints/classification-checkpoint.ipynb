```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

#1 resize images

#first need to load images and convert all to grayscale for Hog
#ford
ford1 = "./ford_train1.jpeg"
ford2 = "./ford_train2.jpeg"
ford3 = "./ford_train3.jpeg"
ford4 = "./ford_train4.png"
ford5 = "./ford_train5.jpeg"

ford1 = cv2.cvtColor(cv2.imread(ford1), cv2.COLOR_BGR2GRAY)
ford2 = cv2.cvtColor(cv2.imread(ford2), cv2.COLOR_BGR2GRAY)
ford3 = cv2.cvtColor(cv2.imread(ford3), cv2.COLOR_BGR2GRAY)
ford4 = cv2.cvtColor(cv2.imread(ford4), cv2.COLOR_BGR2GRAY)
ford5 = cv2.cvtColor(cv2.imread(ford5), cv2.COLOR_BGR2GRAY)



#volkswagen
volks1 = "./volks_train1.png"
volks2 = "./volks_train2.jpeg"
volks3 = "./volks_train3.jpeg"
volks4 = "./volks_train4.jpeg"
volks5 = "./volks_train5.jpeg"

volks1 = cv2.cvtColor(cv2.imread(volks1), cv2.COLOR_BGR2GRAY)
volks2 = cv2.cvtColor(cv2.imread(volks2), cv2.COLOR_BGR2GRAY)
volks3 = cv2.cvtColor(cv2.imread(volks3), cv2.COLOR_BGR2GRAY)
volks4 = cv2.cvtColor(cv2.imread(volks4), cv2.COLOR_BGR2GRAY)
volks5 = cv2.cvtColor(cv2.imread(volks5), cv2.COLOR_BGR2GRAY)

#make lists to store each volks and ford

train_imageslist = [] 
label_imageslist = [] 

#load into array each image with their corresponding label
train_imageslist.append(ford1)
label_imageslist.append(1)
train_imageslist.append(ford2)
label_imageslist.append(1)
train_imageslist.append(ford3)
label_imageslist.append(1)
train_imageslist.append(ford4)
label_imageslist.append(1)
train_imageslist.append(ford5)
label_imageslist.append(1)
train_imageslist.append(volks1)
label_imageslist.append(-1)
train_imageslist.append(volks2)
label_imageslist.append(-1)
train_imageslist.append(volks3)
label_imageslist.append(-1)
train_imageslist.append(volks4)
label_imageslist.append(-1)
train_imageslist.append(volks5)
label_imageslist.append(-1)

#now resize
#iterate through the train images and resize

for i, col in enumerate(train_imageslist):
    index = train_imageslist[i]
    index = cv2.resize(index, (128,128))
    train_imageslist[i] = index

```

```python
#2 compute HoG features of 32x32x8 with block 4x4

#Suggestion: Make a function which takes list of images as 
#arguments and delivers list of HoG features as output. 
#The same function can be used during testing.)

def computehog_features(listofimages):
    
    hoglist = [] #to store our return hog features in function
    cell_size = (4, 4)  # h x w in pixels
    block_size = (4, 4)  # h x w in cells
    nbins = 8  # number of orientation bins because x8

#need to add a For loop here to iterate through each element in the list
    for i, col in enumerate(train_imageslist):
        img = train_imageslist[i]
        # create HoG Object
        # winSize is the size of the image cropped to an multiple of the cell size
        hog = cv2.HOGDescriptor(_winSize=(img.shape[1] // cell_size[1] * cell_size[1],
                                          img.shape[0] // cell_size[0] * cell_size[0]),
                                _blockSize=(block_size[1] * cell_size[1],
                                            block_size[0] * cell_size[0]),
                                _blockStride=(cell_size[1], cell_size[0]),
                                _cellSize=(cell_size[1], cell_size[0]),
                                _nbins=nbins)

        n_cells = (img.shape[0] // cell_size[0], img.shape[1] // cell_size[1])

        # Compute HoG features
        hog_feats = hog.compute(img)\
                       .reshape(n_cells[1] - block_size[1] + 1,
                                n_cells[0] - block_size[0] + 1,
                                block_size[0], block_size[1], nbins) \
                       .transpose((1, 0, 2, 3, 4))  # index blocks by rows first

        # hog_feats now contains the gradient amplitudes for each direction,for each cell of its group for each group.
        # Indexing is by rows then columns.

        # computation for BlockNorm
        gradients = np.full((n_cells[0], n_cells[1], 8), 0, dtype=float)
        cell_count = np.full((n_cells[0], n_cells[1], 1), 0, dtype=int)

        for off_y in range(block_size[0]):
            for off_x in range(block_size[1]):
                gradients[off_y:n_cells[0] - block_size[0] + off_y + 1,
                          off_x:n_cells[1] - block_size[1] + off_x + 1] += \
                    hog_feats[:, :, off_y, off_x, :]
                cell_count[off_y:n_cells[0] - block_size[0] + off_y + 1,
                           off_x:n_cells[1] - block_size[1] + off_x + 1] += 1

        # Average gradients
        gradients /= cell_count
        #now put gradient into list to save 
        hoglist.append(gradients)
    return hoglist


#now use this function to get hog features
hog_trainlist = []
hog_trainlist = computehog_features(train_imageslist)
```

```python
#3 use kneighbor

neigh = KNeighborsClassifier(n_neighbors=3)
#need to convert arrays to numpy
hog_trainlist = np.asarray(hog_trainlist)
hog_trainlist = hog_trainlist.reshape(10,-1)

neigh.fit(hog_trainlist, label_imageslist) 
```

```python
#testing
#resize testing images
fordtest = "./ford_test.jpeg"
fordtest = cv2.cvtColor(cv2.imread(fordtest), cv2.COLOR_BGR2GRAY)
volkstest = "./volks_test.jpeg"
volkstest = cv2.cvtColor(cv2.imread(volkstest), cv2.COLOR_BGR2GRAY)


#create test list to be able to enter into function created above
testlist = []
testlist.append(fordtest)
testlist.append(volkstest)


for i, col in enumerate(testlist):
    index = testlist[i]
    index = cv2.resize(index, (128,128))
    testlist[i] = index



```

```python
#compute hog features
#will use function above with testlist as input and store into new array

hogtest = []
hogtest = computehog_features(testlist)
```
