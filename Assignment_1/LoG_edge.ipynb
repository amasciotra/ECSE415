# Part 2 : Edge Detection

2.2 LOG edge detector


```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#Load image and convert to grayscale
img_woman = "./woman.png"
origwoman = cv2.imread(img_woman,cv2.IMREAD_GRAYSCALE)

#apply gaussian blur to image before computing LAplacian and edges for better result
blurGwoman = cv2.GaussianBlur(origwoman, (5,5), 1.41421356237)

#take laplacian to then find the zero crossings
woman_LoG= cv2.Laplacian(blurGwoman, cv2.CV_32F, ksize = 5)

ht, wt = blurGwoman.shape




#finding zero crossings now, any product that yields a negative answer should be an edge
#either 0 or change of sign when iterating through each pixel

edges = np.zeros((ht, wt), dtype=np.uint8)


for row in range(1, ht-1):
    for col in range(1, wt-1):
       
        
        
        if ((woman_LoG[row,col]*woman_LoG[row-1,col] <= 0) or (woman_LoG[row,col]*woman_LoG[row-1,col] <= 0) or (woman_LoG[row,col]*woman_LoG[row+1,col] <= 0) or (woman_LoG[row,col]*woman_LoG[row,col+1] <= 0)):
            edges [row, col] = 255
        
        else:
                edges[row,col] = 0
                
#plot results

plt.figure(figsize=(15, 15))
plt.subplot(121)
plt.title("Original Woman")
plt.imshow(origwoman, cmap = "gray"),plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.title("Laplacian Edge Detection")
plt.imshow(edges, cmap = "gray"), plt.xticks([]), plt.yticks([])

plt.show()
    










```
