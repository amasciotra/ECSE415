```python
#template matching

#1. with clean Waldo

import cv2
import numpy as np
import matplotlib.pyplot as plt


#load images
img_path = "./where_is_waldo.jpg"
img_scene = cv2.imread(img_path)
img_scene = cv2.cvtColor(img_scene, cv2.COLOR_BGR2RGB)
img_rect = img_scene.copy() #copy for the rectangle
img_rect2 = img_scene.copy() #copy for the rectangle



img_waldotemp = "./template1.png" #clean waldo
temp1 = cv2.imread(img_waldotemp)
temp1 = cv2.cvtColor(temp1, cv2.COLOR_BGR2RGB)


#set size and call templatematch function to set waldo temp

height, width = temp1.shape[0:2]
# apply template matching
# cv2.matchTemplate(image, templ, method) → result
result = cv2.matchTemplate(img_scene, temp1, cv2.TM_SQDIFF)##SSD formula
print(np.min(result)) 

#thresholding will be above 8

#get points above threshold

loc = np.where(result <= 8)

# loop though the points
for pt in zip(*loc[::-1]):
        cv2.rectangle(img_rect, pt, (pt[0]+width, pt[1]+height), (255,255,0), 5)

    
    
# display images
plt.figure(figsize=(15,15))
plt.imshow(img_rect)
plt.title("Matched Image in Yellow square"), plt.xticks([]), plt.yticks([])
plt.show()




```
