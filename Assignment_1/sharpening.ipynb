```python
import numpy as np
import cv2
import matplotlib.pyplot as plt


#load original Rice
img_ricepath = "./rice.png"
origrice = cv2.imread(img_ricepath)

#1.
#blur image with boxfilter 5 x5
blurBrice = cv2.blur(origrice, (5,5))

#blur image with gaussian variance of 2
blurGrice = cv2.GaussianBlur(origrice, (5,5), 1.41421356237)

detailsB = cv2.subtract(origrice, blurBrice) #box
detailsG = cv2.subtract(origrice, blurGrice) #gaussian

#storing final images
finalSharpenedB = cv2.add(origrice, detailsB)
finalSharpenedG = cv2.add(origrice, detailsG)


#plotting
#convert to RGB
origrice           = cv2.cvtColor(origrice, cv2.COLOR_BGR2RGB)
finalSharpenedB    = cv2.cvtColor(finalSharpenedB, cv2.COLOR_BGR2RGB)
finalSharpenedG    = cv2.cvtColor(finalSharpenedG, cv2.COLOR_BGR2RGB)

#1by3 matrix

plt.figure(figsize =(20,20))
plt.subplot(131), plt.imshow(origrice)
plt.title("Original Rice"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(finalSharpenedB)
plt.title("Box Sharpen "), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(finalSharpenedG)
plt.title("Gaussian Sharpen"), plt.xticks([]), plt.yticks([])

plt.show()








```

2. 

The Gaussian Sharpen result is superior to the box sharpen. This is because as mentioned from the denoising result, a gaussian filter is more precise than the box filter, and produces a nicer blur. This is due to the weighted average filter which the Gaussian filter applies vs a straight average filter weighing every cell equally. Since the filter is more precise, this directly translates to the finished image of the Sharpening effect.
