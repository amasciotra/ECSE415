```python
#images were cropped to only include face and resized to 100x100 in Preview(mac) prior to loading into python.


def get_images_in_directory(directory, ext="jpg"):
    """Read the filepaths to all images in a directory.
    Args:
        directory Filepath to the directory.
        ext File extension of the images.
    Returns:
        List of filepaths
    """
    filepaths = []
    for fname in os.listdir(directory):
        if fname.endswith(ext) and os.path.isfile(os.path.join(directory, fname)):
            filepaths.append(os.path.join(directory, fname))
    return filepaths

#loading test image
imgtestpath = "./test/OfficeOlympics.jpg"
imgtest = cv2.cvtColor(cv2.imread(imgtestpath), cv2.COLOR_BGR2RGB)

imgpath = get_images_in_directory("./train/", ext= "jpg") #get a list of all the file paths 

plt.imshow(imgtest) #testing image for later
plt.title("Test image"), plt.xticks([]), plt.yticks([])
plt.show()

#displaying all the test images
train_imgreadcolor = [] #to store images

for i in range(0, len(imgpath)):
    imgtoprint = cv2.cvtColor(cv2.imread(imgpath[i]), cv2.COLOR_BGR2RGB)
    train_imgreadcolor.append(imgtoprint)
    train_imgreadcolor[i] = cv2.resize(train_imgreadcolor[i], (100,100))
    #print(imgpath[i])
#print(imgpath)
#print(train_imgreadcolor)
    plt.imshow(train_imgreadcolor[i])
    plt.title("training images"), plt.xticks([]), plt.yticks([])
    plt.show()
    















```

# 2. Eigen Vector Representation

```python
#first will make a list of grayscale images instead of the images being in color, reduce the RGB channels so only need to flatten 2D

train_imgreadgray = [] #to store images
for i in range(0, len(imgpath)):
    imgtoprint = cv2.cvtColor(cv2.imread(imgpath[i]), cv2.COLOR_BGR2GRAY)
    #imgtoprint = np.float32(imgtoprint)/255.0 #converting to floating point
    #imgtoprint = np.float32(imgtoprint)
    train_imgreadgray.append(imgtoprint)
    train_imgreadgray[i] = cv2.resize(train_imgreadgray[i], (100,100))

    plt.imshow(train_imgreadgray[i], cmap = "gray")
    plt.title("training images"), plt.xticks([]), plt.yticks([])
    plt.show()
    
#make the data matrix to perform PCA on
#each row will be an image flattened, every column will be number of images
datamatrix = np.zeros((len(train_imgreadgray), train_imgreadgray[0].shape[0] * train_imgreadgray[0].shape[1]), dtype=np.float32)
for i in range(0, len(train_imgreadgray)):
#for i in range(20,30):
        img = train_imgreadgray[i].flatten()
        datamatrix[i,:] = img
# print(datamatrix.shape)
# print(len(datamatrix))
# print(datamatrix)

#using sklearn functions 
# Initialize PCA object; use "n_components" to see number of PC
pca = PCA(n_components=30)#lowest number between 30 and 10000 array


# Fit the PCA and return the data using pca.fit_transform( ... )
X_train_pca = pca.fit_transform(datamatrix)

# Print the shape of the transformed data
print(X_train_pca.shape)
print("eigen faces")

# Get the variance explained by every principal component:
varianceplot = pca.explained_variance_ratio_
varianceplotcum=np.cumsum(np.round(varianceplot, decimals=4))


mean = pca.mean_ #calculating the mean image

meanface = mean.reshape(train_imgreadgray[0].shape)#reshaping to be able to display



#now going to get eigenvectors using the pca.components_
eigenvectors = pca.components_

#displaying eigen faces
eigenfaces = []
for i in range(0,30):
    eigenface = eigenvectors[i].reshape(train_imgreadgray[0].shape)
    eigenfaces.append(eigenface)
    plt.imshow(eigenfaces[i], cmap = "gray")
    plt.title("Eigen Faces" + str(i+1)), plt.xticks([]), plt.yticks([])
    plt.show()
    

    
#####this was first attempted usign the opencv library, the SKlearn had better functions to use.


# #computing eigenvectors and mean
# mean2, eigenvectors2 = cv2.PCACompute(datamatrix, mean=None, maxComponents=30)
# print(eigenvectors.shape)
# print(len(eigenvectors))
# print(eigenvectors)
# eigenVectors = sorted(eigenvectors.any(), reverse=True)

#to build eigen faces on top of this
# meanface2 = mean2.reshape(train_imgreadgray[0].shape)
 
# plt.subplot(212).imshow(meanface2, cmap = "gray")
# plt.title("mean image2"), plt.xticks([]), plt.yticks([])
# # #reconstructing eigen face
# eigenfaces = []
# for eigenvector in eigenvectors:
#     eigenface = eigenvector.reshape(train_imgreadgray[0].shape)
#     eigenfaces.append(eigenface)
   


# for i in range(0,7):
#     output = np.add(meanface, eigenfaces[i]*5)
# output = cv2.resize(output, (0,0), fx=2, fy=2)
# plt.subplot(313).imshow(output, cmap = "gray")
# plt.title("output"), plt.xticks([]), plt.yticks([])
# plt.show()




    



```

# 3. Face Detection and Recognition

```python
#recall group image from the first cell is called  imgtest
imgtestcustom = imgtest.copy() #to write rectangles on
imgtestgray = imgtest.copy() #for greyscale
imgtestgray = cv2.cvtColor(imgtestgray, cv2.COLOR_BGR2GRAY) #convert to gray for comparison later

#eigenvectors from above are also eigenvectors, mean called mean
pca = PCA(n_components=25) # taking only 25 best


X_train_pca2 = pca.fit_transform(datamatrix)
mean2 = pca.mean_

eigenvectors2 = pca.components_

window = 100
faces = []
step = 9
thresh = 5000


#sliding window to check if less than threshold, and save these windows
for y in range(0, imgtest.shape[0] - window, step):
    for x in range(0, imgtest.shape[1] - window, step):
        temp = 0
        projection_temp = np.zeros(eigenvectors2.shape[0])
        samplewind = imgtestgray[y:y + window, x:x + window] ##making window on image
        samplewind = cv2.resize(samplewind, (100, 100)) #resize to our train images
        samplewindvector = np.reshape(samplewind, (1, -1))
        
        #reconstructing face in window to see if match
        for i in range(eigenvectors2.shape[0]):
            projection_temp[i] = np.dot((samplewindvector - mean2), eigenvectors2[i]) 
        
        for i in range(eigenvectors2.shape[0]):
            temp += (projection_temp[i]*eigenvectors2[i])
        magnitude = samplewindvector - (temp + mean2)
        value = np.linalg.norm(magnitude)
        
        #comparing value with threshold, filter out lower all big differences to know its a face
        if(value < thresh):
            faces.append(((x,y), value))
            

#only draw one window in one area or else discard
deltafaces = []

for i in range(len(faces)):
    for j in range(len(faces)):
        if(i == j):
            j += 1
            continue
            
        dist = np.subtract(faces[i][0],faces[j][0])
        if(np.hypot(dist[0], dist[1]) < 150):
            if(faces[i][1] >= faces[j][1]):
                deltafaces.append(i)
            else:
                deltafaces.append(j)

duplicate_windows = list(set(deltafaces))
for i in sorted(duplicate_windows, reverse=True):
    del faces[i]




#display the faces
for i in range(len(faces)):
    x, y = faces[i][0][0], faces[i][0][1]
    cv2.rectangle(imgtestcustom, (x, y) ,(x + window, y + window), (255,0,255), 3)
    
plt.figure(figsize=(10,10))
plt.imshow(imgtestcustom)
plt.xticks([]), plt.yticks([])
plt.show()


    
```
